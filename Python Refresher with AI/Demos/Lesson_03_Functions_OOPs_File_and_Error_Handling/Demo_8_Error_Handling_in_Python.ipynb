# **Demo 08: Implementing Error Handling in Python**

**Objective:** To implement error handling in Python using try, except, and finally blocks to manage file operations effectively

**Prerequisites:** None

**Tools required:** Python

### **Steps to be followed:**

1. Prompt the user for a file name
2. Attempt to open and read the file contents
3. Catch the FileNotFoundError if the file does not exist
4. Ensure that the file is always closed, whether or not an error occurred


```python
# Step 1: Prompt the user for a file name
file_name = input("Enter the file name: ")
```

```python
# Step 2: Attempt to open and read the file contents
try:
    with open(file_name, "r") as file:
        # Step 4: Display the file contents if the file is found
        print("\nFile Contents:\n")
        print(file.read())

# Step 3: Catch the FileNotFoundError if the file does not exist
except FileNotFoundError:
    print(f"\nError: The file '{file_name}' was not found.")

finally:
    print("\nProgram has completed execution.")
```

```python
# Step 4: Ensure that the file is always closed (handled automatically by 'with' statement)
print("\nOperation completed.")
```

##### By following these steps, you have successfully learned how to handle errors in Python using try, except, and finally blocks to manage file operations effectively.
