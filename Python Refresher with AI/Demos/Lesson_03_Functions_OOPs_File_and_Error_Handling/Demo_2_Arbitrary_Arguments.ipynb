# **Demo 02: Implementing Arbitrary Arguments**


**Objective:** To demonstrate how to use arbitrary arguments in Python effectively

**Prerequisites:** None

**Tools required:** Python

### **Steps to be followed:**

1. Define a function that takes arbitrary arguments
2. Pass arguments to the function, call the function, and display the results


**Step 1:** Define a function that takes arbitrary arguments

```python
def display_items(*args, **kwargs):
    
    """Function to accept and display arbitrary positional and keyword arguments."""

    # Display positional arguments
    print("Positional Arguments:")
    for index, item in enumerate(args, start=1):
    # for index, item in args: # will not work
        print(f"  Item {index}: {item}")

    # Display keyword arguments
    print("\nKeyword Arguments:")
    for key, value in kwargs.items():
        print(f"  {key}: {value}")
```

**Step 2:** Pass arguments to the function, call the function, and display the results

```python
result_1 = display_items("Apple", "Banana", "Cherry", name="Alice", age=30, city="New York")
```

```python
result_2 = display_items(1, 2, 3, 4, 5, language="Python", version=3.10, stable=True)
```
