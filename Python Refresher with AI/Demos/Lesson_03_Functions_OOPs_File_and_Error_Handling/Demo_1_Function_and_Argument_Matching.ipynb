# **Demo 01: Implementing Function and Argument Matching**

**Objective:** To learn how to create a function with arguments by defining functions, passing parameters, and utilizing arguments effectively

**Prerequisites:** None 

**Tools required:** Python

### **Steps to be followed:**

1. Create a function
2. Call the function and pass arguments to the function
3. Display the results 


**Step 1:** Create a function

```python
def greet(name, greeting="Hello"):
    """Function to greet a person with a custom message."""
    return f"{greeting}, {name}!"
```

**Step 2:** Call the function and pass arguments to the function

```python
name_arg = "Alice"
greeting_arg = "Good morning"
# (Positional argument matching)
result1 = greet(name_arg, greeting_arg)

# (Keyword argument matching)
result2 = greet(name=name_arg, greeting=greeting_arg)

# (Default argument usage)
result3 = greet(name_arg)  # Uses default greeting
```

**Step 3:** Display the results 

```python
print(result1) 
print(result2) 
print(result3) 
```

##### By following these steps, you have successfully learned how to define functions, pass parameters, and effectively utilize arguments in Python.
