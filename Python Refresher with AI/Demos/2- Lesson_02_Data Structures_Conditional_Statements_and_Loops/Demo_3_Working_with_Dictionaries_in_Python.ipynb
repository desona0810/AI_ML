# **Demo 03: Working with Dictionaries in Python**

**Objective:** To create, update, and retrieve data efficiently using Python dictionaries

**Prerequisites:** None

**Tools required:** Python

### **Steps to be followed:**

1. Create a dictionary with at least five key-value pairs, representing a student’s details (e.g., name, age, course, marks)
2. Access and print values using keys
3. Modify an existing value in the dictionary
4. Add a new key-value pair and remove an existing key
5. Retrieve all keys, values, and key-value pairs using dictionary methods
6. Use the get() method to access a key that may or may not exist

**Step 1:** Create a dictionary with at least five key-value pairs, representing a student’s details (e.g., name, age, course, marks)

```python
# 1. Creating a dictionary with at least 5 key-value pairs (student details)
student_dict = {
    "name": "John Doe",
    "age": 21,
    "course": "Computer Science",
    "marks": 85,
    "city": "New York"
}

print(student_dict)
```

**Step 2:** Access and print values using keys

```python
student_name = student_dict["name"]
student_course = student_dict["course"]

print(student_name)
print(student_course)
```

**Step 3:** Modify an existing value in the dictionary

```python
student_dict["marks"] = 90  # Updating marks
print(student_dict)
```

**Step 4:** Add a new key-value pair and remove an existing key

```python
student_dict = {
    "name": "John Doe",
    "age": 21,
    "course": "Computer Science",
    "marks": 85,
    "city": "New York"
}

student_dict["email"] = "johndoe@example.com"  # Adding email
removed_value = student_dict.pop("city")  # Removing city
print(student_dict)
print(removed_value)
```

**Step 5:** Retrieve all keys, values, and key-value pairs using dictionary methods

```python
all_keys = list(student_dict.keys())
all_values = list(student_dict.values())
all_items = list(student_dict.items())

print(all_keys)
print(all_values)
print(all_items)
```

**Step 6:** Use the get() method to access a key that may or may not exist

```python
student_dict = {
    "name": "John Doe",
    "age": 21,
    "course": "Computer Science",
    "marks": 85,
    "city": "New York"
}

existing_value = student_dict.get("course")  # Returns "Computer Science"
non_existing_value = student_dict.get("phone", "Key not found")  # Returns "Key not found"


print(existing_value)
print(non_existing_value)
```
